## Análise por tipo de escola

```
df['tipo_escola'].value_counts()
```

```
df.groupby('tipo_escola')[['nota_matematica','nota_linguagens','nota_ciencias','nota_humanas','nota_redacao','media_final']].mean()
```

## Análise por estado

```
media_estado = df.groupby('estado')['media_final'].mean().sort_values(ascending=False)
media_estado
```

## Ranking estudantes

```
top10 = df.sort_values('media_final', ascending=False).head(10)
top10
```

## Desempenho por disciplina

```
df[['nota_matematica','nota_linguagens','nota_ciencias','nota_humanas','nota_redacao']].mean()
```

## Relação renda x desempenho

```
df[['renda_familiar','media_final']].corr()
```

## 📌 Conclusão


- Escolas privadas tendem a ter desempenho médio maior.
- Alguns estados concentram melhores médias.
- Existe correlação positiva entre renda e desempenho.
- Políticas devem focar em redução de desigualdade educacional.
